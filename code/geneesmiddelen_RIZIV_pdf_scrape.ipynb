```python
def get_all_links():
    '''output all category pagelinks (dig down for pdf's)'''
    url = "https://www.riziv.fgov.be/nl/statistieken/geneesmiddel/Paginas/geneesmiddelen-groep-voorschrijvers.aspx"
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
    name = []
    href = []
    
    for link in soup.select("ul")[5].select("a"):
        #print ("name: {}\nlink:{}\n".format(link.attrs['href'][80:].split(".")[0], link.attrs['href']))
        name.append(link.attrs['href'][80:].split(".")[0])
        href.append(link.attrs['href'])
    return name, href
```

```python
def get_pdfs_from_one_page(href, name):
    '''output all pdf links of that page'''
    url = href
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
    pdf_href_list = []
    
    for link in soup.select("li.dfwp-item"):
        #print(link.select("a")[0].attrs['href'])
        #pdf_href_list.append(link.select("a")[0].attrs['href'][:-4] + "_" + name + ".pdf")
        pdf_href_list.append(link.select("a")[0].attrs['href'])
    return pdf_href_list
```

```python
def load_pdf_stream(link_pdf):
    table_shapes = []
    try:
        tables   = camelot.read_pdf(link_pdf, pages="1-end", flavor='stream', flag_size=True, strip_text=' .\n')
        n_tables = tables.n
        n_pages  = tables[tables.n-1].parsing_report['page']
        for table in range(n_tables - 1):
            table_shapes.append(tables[table].shape)
        return tables-1, str(table_shapes), int(n_tables), int(n_pages), 1
    except:
        return load_pdf_stream_alt1(link_pdf)
```

```python
def load_pdf_stream_alt1(link_pdf):
    table_shapes = []
    try:
        #pdb.set_trace()
        tables   = camelot.read_pdf(link_pdf, pages="1-end", flavor='lattice')
        n_tables = tables.n
        n_pages  = tables[tables.n-1].parsing_report['page']
        tables   = camelot.read_pdf(link_pdf, pages="1-" + str(n_pages - 1), flavor='stream', flag_size=True, strip_text=' .\n')
        for table in range(n_tables - 1):
            table_shapes.append(tables[table].shape)
        return tables-1, str(table_shapes), int(n_tables), int(n_pages), 2
    except:
        return load_pdf_stream_alt2(link_pdf)
```

```python
def load_pdf_stream_alt2(link_pdf):
    return 1, str(1), 1, 1, 3
```

### read the pdf and it can either work or fail

if it works then go to next task
if not then save that path in another failed list

```python
def read_check_pdf(link_pdf):
    try:
        tables = camelot.read_pdf(link_pdf, pages="1-end", flavor='stream', flag_size=True, strip_text=' .\n')
        # return the result, it was success so failed_flag = 0
        failed_flag = 0
        return failed_flag, link_pdf, tables
    except:
        # return the failed pdf link, failed_flag = 1
        failed_flag = 1
        # 1) return the failed or not flag
        # 2) return the original pdf link address to do something with
        return failed_flag, link_pdf, np.nan
```

```python
def read_pdf(failed_flag, link_pdf, tables):
    # here you select if you can grab the head an tail of the 3th table
    count = 0
    if failed_flag == 0:
        n_tables = len(tables)
        for table in range(n_tables):
            n_variables = tables[table].shape[1]
            #print(n_variables)
            if (n_variables == 7):
                x = tables[table].df.loc[0][6]
                if len(x.split(',')) == 1:
                    #print("head of data")
                    count +=1
                #else:
                    #print("tail of data")
        return link_pdf, tables, count
    else:
        return link_pdf, tables, count
        #return failed_flag, link_pdf, tables
```

```python
def transform_store_pdf(link_pdf, tables, count):
    df_list = []
    if count == 1:
        for table in range(len(tables)):
            n_variables = tables[table].shape[1]
            if (n_variables == 7):
                df_list.append(tables[table].df)
        
        df = pd.concat(df_list).reset_index()
        df = df.drop(["index", 0, 1], axis=1)
        df["jaar"] = link_pdf.split("_")[4].split(".")[0]
        df["specialisten"] = link_pdf.split("_")[3]
        df.columns = ["atc", "werkzaam_bestanddeel", "bruto", "bruto_pct", "aantal_DDD", "jaar", "specialisten"]
        #pdb.set_trace()
        for col in df["bruto_pct"]:
            if (len(col.split(',')) == 1):
                df = df.drop([0], axis=0).reset_index()
                df = df.drop(["index"], axis=1)
        return df, "done"
    else:
        # return an empty dataframe
        return pd.DataFrame(columns=["atc", "werkzaam_bestanddeel", "bruto", "bruto_pct", "aantal_DDD", "jaar", "specialisten"]), link_pdf
```

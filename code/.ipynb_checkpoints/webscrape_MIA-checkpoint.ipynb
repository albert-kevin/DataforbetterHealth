now we can start creating a function to get the data from 1 line

```python
def get_single_line(li):
    '''returns the yellow and/or white data from one line.

        li    : soup.select("li")[i]
                this means i is on of the 12 lines on webpage
                
        output: if 3 elements return list[code, description]
                if 2 elements return list[description]'''
    
    # remove all white spaces
    unique = set(li)
    unique.remove(' ')
    
    # check for 3 elements (red, yellow, white)
    if len(unique) == 3:
        yellow = li.select("a")[0].text.strip()
        white  = li.select("a")[0].next_sibling.strip()
        return [yellow, white]
    
    # check for 2 elements (red, white)
    if len(unique) == 2:
        white  = li.contents[2].strip()
        return [white]
```

```python
# test function
for i in range(12):
    print (get_single_line(li=soup.select("li")[i]))
```

now we can start a function to get the data from 1 page

```python
def get_single_page(soup):
    
    '''returns the yellow and/or white data from one page.
    using function "get_single_line()".'''
    
    data_list = [] #make an empty list
    
    # select each of the 12 lines of one page and stack
    data_list = [get_single_line(li=soup.select("li")[line_nr]) for line_nr in range(12)]
    return list(itertools.chain.from_iterable(data_list)) # transform into a single list
```

```python
dataframe_column_list = ["atc_code_1", "atc_descr_1",
                         "atc_code_2", "atc_descr_2",
                         "atc_code_3", "atc_descr_3",
                         "atc_code_4", "atc_descr_4",
                         "atc_code_5", "atc_descr_5",
                         "chemical_name",
                         "chemical_smi",
                         "icd_code_5", "icd_descr_5",
                         "icd_code_4", "icd_descr_4",
                         "icd_code_3", "icd_descr_3",
                         "icd_code_2", "icd_descr_2",
                         "icd_code_1", "icd_descr_1"]
df = pd.DataFrame(columns=dataframe_column_list)
```

#### Now let's start creating the function to get all pages

```python
# initialize the columns (empty dataset)
dataframe_column_list = ["atc_code_1", "atc_descr_1",
                         "atc_code_2", "atc_descr_2",
                         "atc_code_3", "atc_descr_3",
                         "atc_code_4", "atc_descr_4",
                         "atc_code_5", "atc_descr_5",
                         "chemical_name",
                         "chemical_smi",
                         "icd_code_5", "icd_descr_5",
                         "icd_code_4", "icd_descr_4",
                         "icd_code_3", "icd_descr_3",
                         "icd_code_2", "icd_descr_2",
                         "icd_code_1", "icd_descr_1"]
df = pd.DataFrame(columns=dataframe_column_list)

# go to each of the 3420 pages (!! 3421 exclusive 1)
for page_nr in range(3421):
    url = "http://hulab.rxnfinder.org/mia/miaList/?page=" + str(page_nr)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
    
    # scrape a single webpage and create 1 dataframe
    one_page_df = pd.DataFrame([get_single_page(soup)], columns=dataframe_column_list)
    
    # add this dataframe as one row in the final dataset
    df = df.append(one_page_df, ignore_index=True)
    print ("page nr:{}/{}".format(page_nr, len(df)))

# expected number of records = 3420
len(df)
```

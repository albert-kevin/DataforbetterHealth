```python
# https://docs.python.org/3.4/library/xml.etree.elementtree.html
import xml.etree.cElementTree as ET

import pandas as pd
import re
import pdb
```

```python
#load data file
file = 'REF-1551442403585.xml'
tree = ET.ElementTree(file=file)
df = pd.DataFrame()
```

```python
# initialise
pattern = r"[A-Z]+\w+"
#df = pd.DataFrame()
rowindex = 0

# child element
root = tree.getroot()
#pdb.set_trace()
for AtcClassification in root:
    #check for selecting the correct tag:
    #pdb.set_trace()
    if (AtcClassification.tag == '{urn:be:fgov:ehealth:samws:v2:refdata}AtcClassification'):
        #pdb.set_trace()
        # define the attribute name
        df.loc[rowindex, "ATC_Code"] = AtcClassification.get('code')
        #df.loc[rowindex, "Ingredient_CodeType"] = AtcClassification.get('codeType')
        for attr in AtcClassification:
            #pdb.set_trace()
            # get attributes rank etc
            # use print(attr.attrib) to read all
            #df.loc[rowindex, "Ingredient_Synonym_Lang"] = attr.get('{http://www.w3.org/XML/1998/namespace}lang')
            #lang = attr.get('Description')
            #df.loc[rowindex, "Ingredient_Synonym_Rank"] = attr.get('rank')
            #rank = attr.get('rank')
            #pdb.set_trace()
            temp = "ATC_Code_" + re.findall(pattern, attr.tag)[0]
            df.loc[rowindex, temp] = attr.text
        # below must be in the Data loop
        rowindex += 1
```

```python
# save the dataframe to csv
df.to_csv("output/ATC_code"+file[:-3]+"csv", index=False)
```

```python
# initialise
pattern = r"[A-Z]+\w+"
df = pd.DataFrame()
rowindex = 0

# child element
root = tree.getroot()
#pdb.set_trace()
for DeliveryModus in root:
    #rowindex += 1
    #check for selecting the correct tag:
    #pdb.set_trace()
    
    if (DeliveryModus.tag == '{urn:be:fgov:ehealth:samws:v2:refdata}DeliveryModus'):
        #pdb.set_trace()
        # define the attribute name
        df.loc[rowindex, "ATC_Delivery_Modus"] = DeliveryModus.get('code')
        #df.loc[rowindex, "Ingredient_CodeType"] = AtcClassification.get('codeType')
        for attr in DeliveryModus:
            #pdb.set_trace()
            # get attributes rank etc
            # use print(attr.attrib) to read all
            df.loc[rowindex, "Ingredient_Synonym_Lang"] = attr.get('{http://www.w3.org/XML/1998/namespace}lang')
            #lang = attr.get('Description')
            #df.loc[rowindex, "Ingredient_Synonym_Rank"] = attr.get('rank')
            #rank = attr.get('rank')
            #pdb.set_trace()
            temp = "ATC_Delivery_Modus_" + re.findall(pattern, attr.tag)[0]
            for lang in attr:
                #pdb.set_trace()
                temp2 = temp + "_" + re.findall(pattern, lang.tag)[0]
                df.loc[rowindex, temp2] = lang.text
        # below must be in the Data loop
        rowindex += 1
```

```python
# save the dataframe to csv
df.to_csv("output/ATC_Delivery_Modus"+file[:-3]+"csv", index=False)
```

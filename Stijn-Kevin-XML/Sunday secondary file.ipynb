```python
# https://docs.python.org/3.4/library/xml.etree.elementtree.html
import xml.etree.cElementTree as ET

import pandas as pd
import re
import pdb
```

```python
#load data file
file = 'CMP-1551442404739.xml'
tree = ET.ElementTree(file=file)
```

```python
# initialise
pattern = r"[A-Z]+\w+"
df = pd.DataFrame()
rowindex = 0

# child element
root = tree.getroot()

for CompoundingIngredient in root:
    #check for selecting the correct tag:
    if (CompoundingIngredient.tag == '{urn:be:fgov:ehealth:samws:v2:export}CompoundingIngredient') or \
    (CompoundingIngredient.tag == '{urn:be:fgov:ehealth:samws:v2:export}CompoundingFormula'):
        # define the attribute name
        df.loc[rowindex, "Ingredient_Code"] = CompoundingIngredient.get('code')
        df.loc[rowindex, "Ingredient_CodeType"] = CompoundingIngredient.get('codeType')
    for Data in CompoundingIngredient:
        if (Data.tag == '{urn:be:fgov:ehealth:samws:v2:export}Data'):
            df.loc[rowindex, "From_Date"] = Data.get('from')
            # now we add the attributes of this level:
            for attr in Data:
                # get attributes rank etc
                # use print(attr.attrib) to read all
                #df.loc[rowindex, "Ingredient_Synonym_Lang"] = attr.get('{http://www.w3.org/XML/1998/namespace}lang')
                lang = attr.get('{http://www.w3.org/XML/1998/namespace}lang')
                #df.loc[rowindex, "Ingredient_Synonym_Rank"] = attr.get('rank')
                rank = attr.get('rank')
                #pdb.set_trace()
                temp = re.findall(pattern, attr.tag)[0] + "_" + lang + "_" + rank
                df.loc[rowindex, temp] = attr.text
        # below must be in the Data loop
        rowindex += 1
```

```python
# save the dataframe to csv
df.to_csv("output/"+file[:-3]+"csv", index=False)
```
